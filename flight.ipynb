```python
import seaborn as sns
import matplotlib.pyplot as plt

source_city_order = ["Delhi", "Mumbai", "Bangalore","Kolkata","Hyderabad","Chennai"]
destination_city_order = ["Delhi", "Mumbai", "Bangalore","Kolkata","Hyderabad","Chennai"]

fig,ax = plt.subplots(1, 2, figsize=(20, 6))
sns.barplot(x="source_city", y="price", data=df, ax=ax[0], order=source_city_order)
sns.barplot(x="destination_city", y="price", data=df, ax=ax[1], order=destination_city_order)
plt.show()
```

```python
#visualization of categoric features with countplot
plt.figure(figsize=(15,23))

plt.subplot(4, 2, 1)
sns.countplot(x=df["airline"], data=df)
plt.title("Frequency of Airline")
;
plt.subplot(4, 2, 2)
sns.countplot(x=df["source_city"], data=df)
plt.title("Frequency of Source City")

plt.subplot(4, 2, 3)
sns.countplot(x=df["departure_time"], data=df)
plt.title("Frequency of Departure Time")

plt.subplot(4, 2, 4)
sns.countplot(x=df["stops"], data=df)
plt.title("Frequency of Stops")

plt.subplot(4, 2, 5)
sns.countplot(x=df["arrival_time"], data=df)
plt.title("Frequency of Arrival Time")

plt.subplot(4, 2, 6)
sns.countplot(x=df["destination_city"], data=df)
plt.title("Frequency of Destination City")

plt.subplot(4, 2, 7)
sns.countplot(x=df["class"], data=df)
plt.title("Class Frequency")

plt.show()

```
